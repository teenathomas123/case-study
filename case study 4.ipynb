# Check whether all the zones generate the same amount of sales.

Null Hypothesis is the mean of the sales generated by the four zones are equal.
Alternate Hypothesis is the mean of the sales generated by the four zones are not equal.

```python
f,p=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])

print('F-statistic=',f)
print('p-value=',p)
```

```python
if p<0.05:
    print('Reject null Hypothesis.')
else:
    print('Accept null hypothesis.')
```

that means the mean of the sales generated by the four zones are not equal.
